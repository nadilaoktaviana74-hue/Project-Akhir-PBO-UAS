```python
class Task:
    def __init__(self, title):
        # Encapsulation
        self.__title = title
        self.__completed = False

    # Getter
    def get_title(self):
        return self.__title

    def is_completed(self):
        return self.__completed

    # Setter
    def set_title(self, title):
        self.__title = title

    # Method
    def mark_completed(self):
        self.__completed = True

    # Polymorphism Method
    def display_info(self):
        status = "Selesai" if self.__completed else "Belum"
        print(f"[Task] {self.__title} | Status: {status}")
```

```python
class DeadlineTask(Task):
    def __init__(self, title, deadline):
        super().__init__(title)
        self.deadline = deadline

    # Override Method
    def display_info(self):
        status = "Selesai" if self.is_completed() else "Belum"

        print(
            f"[Deadline Task] {self.get_title()} "
            f"| Deadline: {self.deadline} "
            f"| Status: {status}"
        )
```

```python
class PriorityTask(Task):
    def __init__(self, title, priority):
        super().__init__(title)
        self.priority = priority

    # Override Method
    def display_info(self):
        status = "Selesai" if self.is_completed() else "Belum"

        print(
            f"[Priority Task] {self.get_title()} "
            f"| Priority: {self.priority} "
            f"| Status: {status}"
        )
```

```python
class TaskManager:
    def __init__(self):
        self.tasks = []

    # Menambahkan task
    def add_task(self, task):
        self.tasks.append(task)

    # Menampilkan semua task
    def show_tasks(self):
        print("\n===== DAFTAR TUGAS =====")

        if not self.tasks:
            print("Belum ada tugas.")
            return

        for index, task in enumerate(self.tasks, start=1):
            print(f"{index}. ", end="")
            task.display_info()

    # Menandai task selesai
    def complete_task(self, index):
        if 0 <= index < len(self.tasks):
            self.tasks[index].mark_completed()
            print("Tugas berhasil diselesaikan.")
        else:
            print("Index tidak valid.")

    # Menghapus task
    def delete_task(self, index):
        if 0 <= index < len(self.tasks):
            deleted = self.tasks.pop(index)
            print(f"Tugas '{deleted.get_title()}' berhasil dihapus.")
        else:
            print("Index tidak valid.")
```

```python
def main():
    manager = TaskManager()

    while True:
        print("\n===== TO-DO LIST =====")
        print("1. Tambah Task Biasa")
        print("2. Tambah Deadline Task")
        print("3. Tambah Priority Task")
        print("4. Tampilkan Semua Task")
        print("5. Selesaikan Task")
        print("6. Hapus Task")
        print("7. Keluar")

        choice = input("Pilih menu: ")

        # Tambah Task Biasa
        if choice == "1":
            title = input("Masukkan judul task: ")
            task = Task(title)
            manager.add_task(task)

        # Tambah Deadline Task
        elif choice == "2":
            title = input("Masukkan judul task: ")
            deadline = input("Masukkan deadline: ")

            task = DeadlineTask(title, deadline)
            manager.add_task(task)

        # Tambah Priority Task
        elif choice == "3":
            title = input("Masukkan judul task: ")
            priority = input("Masukkan prioritas (High/Medium/Low): 3")

            task = PriorityTask(title, priority)
            manager.add_task(task)

        # Tampilkan Task
        elif choice == "4":
            manager.show_tasks()

        # Selesaikan Task
        elif choice == "5":
            manager.show_tasks()

            index = int(input("Pilih nomor task: ")) - 1
            manager.complete_task(index)

        # Hapus Task
        elif choice == "6":
            manager.show_tasks()

            index = int(input("Pilih nomor task yang dihapus: ")) - 1
            manager.delete_task(index)

        # Keluar
        elif choice == "7":
            print("Program selesai.")
            break

        else:
            print("Pilihan tidak valid.")


# Menjalankan Program
main()
```
